```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df = df.drop(['id', 'Product ID'], axis=1)
```

```python
set(df['Type'])
# l - 0
# m - 1
# h - 2
```

```python
def typeMapping(t):
    if t == 'L': return 0
    elif t == 'M': return 1
    else: return 2
```

```python
df.head()
df['Type'] = df['Type'].map(typeMapping)
```

```python
# df.dropna(inplace=True)
# df.head()
# df.tail()
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import numpy as np
```

```python
X = np.array(df.drop(['Machine failure'], axis=1))
y = np.array(df['Machine failure'])
```

```python
X = preprocessing.scale(X)

df.dropna(inplace=True)
y = np.array(df['Machine failure'])
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf = svm.SVC()
clf.fit(X_train, y_train)
```

```python
accuracy = clf.score(X_test, y_test)
accuracy
```
