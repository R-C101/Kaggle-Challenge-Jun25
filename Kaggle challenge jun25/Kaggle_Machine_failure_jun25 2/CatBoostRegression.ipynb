### Using Kernal Ridge Regression

```python
import numpy as np
import pandas as pd
from catboost import CatBoostRegressor
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
df = df.drop(['id', 'Product ID'], axis=1)
```

```python
df.head()
```

```python
def typeToInt(t):
    if t == 'L':
        return 0
    elif t == 'M':
        return 1
    else:
        return 2
```

```python
df['Type'] = df['Type'].map(typeToInt)
```

```python
df.head()
```

```python
X = np.array(df.drop(['Machine failure'], axis=1))
y = np.array(df['Machine failure'])
```

```python
# cb = CatBoostRegressor(learning_rate=0.08899863)
cb = CatBoostRegressor(diffusion_temperature=0.7,learning_rate=0.085,random_seed=12)
cb.fit(X, y)
```

## Preditcting using catboost regressor

```python
test_df = pd.read_csv('test.csv')
test_df.head()
```

```python
test_df['Type'] = test_df['Type'].map(typeToInt)
```

```python
predictions = cb.predict(np.array(test_df.drop(['id', 'Product ID'], axis=1)))
```

```python
test_df['Machine failure'] = predictions
```

```python
test_df.head()
```

```python
test_df.filter(['id', 'Machine failure'], axis=1).to_csv("7_kaggle_machine_failure_pred.csv", index=False)
```
