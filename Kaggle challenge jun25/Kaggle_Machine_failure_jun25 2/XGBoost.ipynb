```python
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
def typeToInt(t):
    if t == 'L':
        return 0
    elif t == 'M':
        return 1
    else:
        return 2
```

```python
df.dropna(inplace=True)
df = df.drop(['id', 'Product ID'], axis=1)
df['Type'] = df['Type'].map(typeToInt)
```

```python
df.head()
```

```python
df.head()
```

```python
X = np.array(df.drop(['Machine failure'], axis=1))
y = np.array(df['Machine failure'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
xgb = XGBRegressor()
xgb.fit(X_train, y_train)
```

```python
xgb.score(X_test, y_test)
```

```python
cat = CatBoostRegressor()
cat.fit(X_train, y_train)
```

```python
cat.score(X_test, y_test)
```

```python
sgd = SGDRegressor()
sgd.fit(X_train, y_train)
```

```python
sgd.score(X_test, y_test)
```

```python
cat2 = CatBoostRegressor(random_seed=42, verbose=150)
cat2.fit(X_train, y_train)
```

```python
cat2.score(X_test, y_test)
```

## Fitting the entire data on cat 2

```python
cat2.fit(X, y)
```

```python
test_df = pd.read_csv('test.csv')
```
